```
#Predykcja dla Polski
#Przy u≈ºyciu Analizy szereg√≥w czasowych - model SARIMAX
Poland = df[df['countriesAndTerritories']=='Poland']
Poland = Poland[['dateRep','cases']]
Poland['dateRep'] = pd.to_datetime(Poland['dateRep'],format='%d/%m/%Y')
Poland = Poland.sort_values(by='dateRep',ascending=True)
Poland = Poland.set_index('dateRep')
```

```
rcParams['figure.figsize'] = 18, 8

#Dekompozycja
decomposition = sm.tsa.seasonal_decompose(Poland, model='additive',freq=30)
fig = decomposition.plot()
plt.show()
```

```
#Wybieranie optymalnych parametr√≥w p,d,q dla modelu SARIMAX

p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```
#Wybieranie optymalnych parametr√≥w p,d,q dla modelu SARIMAX
for param in pdq:
  for param_seasonal in seasonal_pdq:
    try:
      mod = sm.tsa.statespace.SARIMAX(Poland,
                                      order=param,
                                      seasonal_order=param_seasonal,
                                      enforce_stationarity=False,
                                      enforce_invertibility=False)
      results = mod.fit()
      
      print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
    except:
      continue
```

```
#Optymalne parametry dla modelu: ARIMA(1, 1, 1)x(0, 1, 1, 12)12 - AIC:535.5900203180685

mod = sm.tsa.statespace.SARIMAX(Poland,
                                order=(1, 1, 1),
                                seasonal_order=(0, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```
#Analiza reszt modelu
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```
